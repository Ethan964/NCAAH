# Visualizing Difference of Play Style Per Conference

+ We've queried all the appropriate data from Sport Logiq
+ Concatted every team in the nation into one master dataframe and added each teams respective conference
+ Conducted tukey tests on every metric to see if there is a statistically significant difference by conference
+ exported into 'ss_cols_df.csv'
+ now we are perfectly set up to visualize some key stats

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/longer/uconn/data/ss_cols_df.csv')



sns.set_theme(style='whitegrid')
metric = 'Dump In Rate'
plt.figure(figsize=(10,6))

sns.boxplot(data=df, x='Conference', y=metric, hue='Conference', palette='Set2', showfliers=False)

sns.swarmplot(data=df, x='Conference', y=metric, color='0.25', alpha=0.7, size=6)

plt.title(f'Distribution of {metric} by Conference', fontsize=16, fontweight='bold')
plt.ylabel('Standardized Score')
plt.xlabel('Conference')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# creating function for comparing statistics

def conference_comparison(metric):
    sns.boxplot(data=df, x='Conference', y=metric, hue='Conference', palette='Set2', showfliers=False)

    sns.swarmplot(data=df, x='Conference', y=metric, color='0.25', alpha=0.7, size=6)

    plt.title(f'Distribution of {metric} by Conference', fontsize=16, fontweight='bold')
    plt.ylabel('Standardized Score')
    plt.xlabel('Conference')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

conference_comparison('Total NZ Undisciplined Penalties')
```

```python
conference_comparison('Successful Open Ice Dekes')
```

# Quick Hits: Conference Play Style Takeaways

**Dump-In Rates**

+ Hockey East shows the greatest dispersion of dump in rates out of any conference
+ Big Ten has the lowest central tendency of dump in rate, which means they as a conference prioritize more puck possession
+ The Independent teams, show a great tendency of dumping the puck in, beware small sample size
+ CCHA and the AHA shows the highest tendency to dump the puck in with outliers near the top of the graph
    + Shows a focus on discipline, fore-checking pressure, and tactical systems

**Undisciplined Penalty Rates**

+ The Big Ten has the highest tendency to take undicsiplined penalties in the NZ sporting a fairly wide dispersion
+ The five independent teams are also extremely undisciplined
+ Hockey East sports some of the most discplined and undiscplined teams in D1 with wide spanning whiskers
+ ECAC puts extreme importance on discipline in the NZ

**Successful Open Ice Dekes**

+ AHA has the highest dispersion of individual talent/coaching philosophy, has teams 2 Std. Devs above and a team -2 Std. Devs below
+ Big Ten unsurprisingly has the highest amoount of successful open ice dekes per conference
+ Independent teams and ECAC has the least amount of successful dekes which aligns with what we've seen in penalty metrics
