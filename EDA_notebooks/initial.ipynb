```python
merged_df = merged_df.fillna(0)

merged_df.columns


```

```python
numeric_cols = [
    'ES IND XGFP20',
    'ES ACT2XPF',
    'SOO',
    'SOOdiff',
    'SOT%',
    'XGF %',
    'ES CORSIFOR%'
]

merged_df[numeric_cols] = merged_df[numeric_cols].apply(
    pd.to_numeric, errors='coerce'
)

percent_cols = [
    'SOT%',
    'ES OZ F/O Win %',
    'XGF %'
]

merged_df[percent_cols] = (
    merged_df[percent_cols]
    .replace('%', '', regex=True)
    .apply(pd.to_numeric, errors='coerce')
)

merged_df[percent_cols] = merged_df[percent_cols] / 100
merged_df = merged_df.round(2)


merged_df
```

```python
df_prep = merged_df.copy()

df_prep = df_prep[df_prep['Player Total Games Played'] >= 10]

df_prep = df_prep.drop(columns = ['ES CORSIFOR%'])

df_prep['Shots_per_xG'] = df_prep['ES ISF (P20 WOI)'] / (df_prep['ES IND XGFP20'] + 1e-9)

df_prep['Shots_per_xG'] = df_prep['Shots_per_xG'].round(2)
```

```python
cluster_features = [
    'eastOuterSlot-shotsTotalPercentage', 
    'innerSlot-shotsTotalPercentage',
    'outsideNorthEast-shotsTotalPercentage',
    'outsideNorthWest-shotsTotalPercentage',
    'westOuterSlot-shotsTotalPercentage',
    'Shots_per_xG' #
]

df_cluster_subset = df_prep.set_index('Player')[cluster_features]

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_array =  scaler.fit_transform(df_cluster_subset)

scaled_df = pd.DataFrame(
    scaled_array,
    index = df_cluster_subset.index,
    columns = df_cluster_subset.columns
)

scaled_df.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_components = pca.fit_transform(scaled_df)

pca_df = pd.DataFrame(
    data=pca_components,
    columns = ['PCA1', 'PCA2'],
    index = scaled_df.index
)

print(f'Explained Variance Ratio by Component: {pca.explained_variance_ratio_}')
print(f'Total Variance Explained: {sum(pca.explained_variance_ratio_) * 100:.2f}%\n')
```

```python
inertias = []

max_k = min(10, len(pca_df))

for k in range(1, max_k):
    kmeans = KMeans(n_clusters=k, random_state=617, n_init=10)
    kmeans.fit(pca_df)
    inertias.append(kmeans.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(range(1, max_k), inertias, marker='o', linestyle='--', color='b')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.xticks(range(1, max_k))
plt.grid(True, alpha=0.5)
plt.show()
```

```python
k = 3

kmeans = KMeans(n_clusters=3, random_state=617, n_init=10)

pca_df['Cluster'] = kmeans.fit_predict(pca_df)

plt.figure(figsize=(10,7))
for cluster in range(3):
    cluster_subset = pca_df[pca_df['Cluster'] == cluster]
    plt.scatter(
        cluster_subset['PCA1'],
        cluster_subset['PCA2'],
        s=120,
        label=f'Cluster {cluster}',
        edgecolors='black'
    )
for player_name in pca_df.index:
    plt.annotate(
        player_name, 
        (pca_df.loc[player_name, 'PCA1'], pca_df.loc[player_name, 'PCA2']),
        textcoords="offset points", 
        xytext=(0, 8),
        ha='center',
        fontsize=9
    )

plt.title('K-Means Clustering of UConn Skaters (PCA Reduced)', fontsize=14)
plt.xlabel('Principal Component 1 (Playstyle Variance)', fontsize=12)
plt.ylabel('Principal Component 2 (Secondary Variance)', fontsize=12)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.2)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.2)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
loadings_df = pd.DataFrame(
    pca.components_.T,
    columns=['PC1 Weight', 'PC2 Weight'],
    index=cluster_features
    )

loadings_df.sort_values(by='PC2 Weight', ascending=False).round(2)

print(loadings_df)
```
